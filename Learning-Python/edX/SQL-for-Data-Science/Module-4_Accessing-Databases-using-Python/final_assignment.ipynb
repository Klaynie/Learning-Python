### Problem 1
Find the total number of crimes recorded in the CRIME table.

```python
query = """
SELECT
    COUNT(*)
FROM
    CHICAGO_CRIME_DATA;
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 2
List community areas with per capita income less than 11000.

```python
query = """
SELECT
    *
FROM
    CENSUS_DATA
WHERE
    PER_CAPITA_INCOME < 11000;
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 3

List all case numbers for crimes involving minors

```python
query = """
SELECT
    *
FROM
    CHICAGO_CRIME_DATA
WHERE
    DESCRIPTION LIKE '%MINOR%';
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 4

List all kidnapping crimes involving a child (children are not considered minors for the purposes of crime analysis)

```python
query = """
SELECT
    *
FROM
    CHICAGO_CRIME_DATA
WHERE
    PRIMARY_TYPE = 'KIDNAPPING'
AND
    DESCRIPTION LIKE '%CHILD%';
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 5
What kind of crimes were recorded at schools?

```python
query = """
SELECT
    DISTINCT(PRIMARY_TYPE)
FROM
    CHICAGO_CRIME_DATA
WHERE
    LOCATION_DESCRIPTION LIKE '%SCHOOL%'
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 6

List the average safety score for all types of schools.

```python
cursor.execute("UPDATE CHICAGO_PUBLIC_SCHOOLS SET SAFETY_SCORE = NULL WHERE SAFETY_SCORE = '';")
cnx.commit()

query = """
SELECT
    SCHOOL_TYPE,
    AVG(CAST(SAFETY_SCORE AS UNSIGNED)) AS AVG_SAFETY_SCORE
FROM
    CHICAGO_PUBLIC_SCHOOLS
WHERE
    SAFETY_SCORE IS NOT NULL
GROUP BY
    SCHOOL_TYPE
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 9

Use a sub-query to find the name of the community area with highest hardship index.

```python
query = """
SELECT
    COMMUNITY_AREA_NAME
FROM
    CENSUS_DATA
WHERE
    HARDSHIP_INDEX = (
        SELECT
            MAX(HARDSHIP_INDEX)
        FROM
            CENSUS_DATA
        );
"""

df = pd.read_sql(query, con=cnx)
df
```

### Problem 10

Use a sub-query to determine the Community Area Name with most number of crimes

```python
query = """
SELECT
    COMMUNITY_AREA_NAME
FROM
    CENSUS_DATA 
WHERE
    COMMUNITY_AREA_NUMBER = (
        SELECT
            COMMUNITY_AREA_NUMBER
        FROM
            CHICAGO_CRIME_DATA
        GROUP BY
            COMMUNITY_AREA_NUMBER
        ORDER BY
            COUNT(*) DESC
        LIMIT
            1
    );
"""

df = pd.read_sql(query, con=cnx)
df
```
