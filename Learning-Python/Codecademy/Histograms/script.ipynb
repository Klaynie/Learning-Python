# Histograms

Use the `np.histogram()` function to determine the busiest six hours of the day. Save the result to `times_hist`.

Use the following range and number of bins.

* Range: 0 to 24
* Bins: 4

```python
# Import packages
import numpy as np
import pandas as pd

# Read in transactions data
transactions = pd.read_csv("transactions.csv")

# Save transaction times to a separate numpy array
times = transactions["Transaction Time"].values

# Use numpy.histogram() below
times_hist = np.histogram(times, range=(0,24), bins=4)

print(times_hist)
```

## Plotting a Histogram

Use the `plt.hist()` function to create a plot for each six-hour period in a day. Use the following range and number of bins.

* Range: 0 to 24
* Bins: 4

```python
# import pyplot as plt
from matplotlib import pyplot as plt

# Read in transactions data
transactions = pd.read_csv("transactions.csv")

# Save transaction times to a separate numpy array
times = transactions["Transaction Time"].values

# Use plt.hist() below
plt.hist(times, range =(0,24), bins=4, edgecolor='black')
 
plt.title("Customer Frequency")
plt.xlabel("Time")
plt.ylabel("Count")
 
plt.show()
```

This histogram is helpful for our store manager. The last six hours of the day are the busiest — from 6 pm until midnight. Does this mean the manager should staff their grocery store with the most employees between 6 pm and midnight?

To the manager, this does not make much sense. The manager knows the store is busy when many people get off work, but the rush certainly does not continue later than 9 pm.

The issue with this histogram is that we have too few bins. When plotting a histogram, it is essential to select bins that fully capture the trends in the underlying data. Often, this will require some guessing and checking. There is not much of a science to selecting bin size.

How many bins do you think makes sense for this example? I would try 24 because there are 24 hours in a day.

```python
# Import packages
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# Read in transactions data
transactions = pd.read_csv("transactions.csv")

# Save transaction times to a separate numpy array
times = transactions["Transaction Time"].values

# Use plt.hist() below
plt.hist(times, range=(0, 24), bins=24,  edgecolor="black")
plt.title("Weekday Frequency of Customers")
plt.xlabel("Hours (1 hour increments)")
plt.ylabel("Count")

plt.show()
```

<i>Hint: It looks like the busiest times of day are in the morning, from 5am to 10am, and in the evening from 5pm to 10pm.</i>

<i>This histogram has two distinct peaks, neither of which are close to our average of 3pm. As you can see, averages do not tell the full story. By visualizing the shape of our data, we can make better-informed decisions.</i>
