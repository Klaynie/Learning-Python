```python
from matplotlib import pyplot as plt

past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]
```

<img src="https://content.codecademy.com/courses/matplotlib/project_ii/sidebyside.png" alt="side by side bar graph" style="background-color:white;"></img><br>

```python
from matplotlib import pyplot as plt

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
middle_school_a = [80, 85, 84, 83, 86]
middle_school_b = [73, 78, 77, 82, 86]

def create_x(t, w, n, d):
    return [t*x + w*n for x in range(d)]
```

1. Using `create_x`, make the lists `school_a_x` and `school_b_x` which will determine where to put the bars for Middle School A and Middle School B along the x-axis.

```python
school_a_x = create_x(2, 0.8, 1, len(middle_school_a))
school_b_x = create_x(2, 0.8, 2, len(middle_school_b))
```

2. Create a figure of width `10` and height `8`.

```python
%%capture
plt.figure(figsize=(10,8))
```

3. Create a set of axes and save them to `ax`.

```python
%%capture
ax = plt.subplot()
```

```python
plt.figure(figsize=(10,8))
plt.bar(school_a_x, middle_school_a)
plt.bar(school_b_x, middle_school_b)
ax = plt.subplot()
plt.show()
```

5. Create a new list of x-values called `middle_x`, which are the values in the middle of `school_a_x` and `school_b_x`. This is where we will place the x-ticks. Look at the final graph to see this placement.

```python
middle_x = [round((a + b) / 2, 2) for a, b in zip(school_a_x, school_b_x)]
print(school_a_x)
print(school_b_x)
print(middle_x)
```

## Stacked Bars

Now, we are going to look at the chart called `stacked-bars.png`. This graph displays the breakdown of students who got `As`, `Bs`, `Cs`, `Ds`, and `Fs` in each unit.<br>
<br>
<img src="https://content.codecademy.com/courses/matplotlib/project_ii/stacked.png" alt="stacked bar chart" style="background-color:white;"></img><br>
<br>
The data you will need to recreate this chart is in the lists `As`, `Bs`, `Cs`, `Ds`, `Fs`, and `unit_topics`.<br>
<br>
Save your recreated chart to a file called `my_stacked_bar.png`.

```python
from matplotlib import pyplot as plt
import numpy as np

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)

c_bottom = np.add(As, Bs)
```

1. The `Bs` bars will go on top of the `As` bars, but at what heights will the `Cs`, `Ds`, and `Fs` bars start?<br>
<br>
The bottom of the bars representing the `Cs` will be at the height of the `As` plus the `Bs`. We can do this in NumPy (a scientific computing package for Python) with the `np.add` function. `c_bottom`, the starting heights for the `Cs`, will be:

    c_bottom = np.add(As, Bs)

* Define `d_bottom` (where the `Cs` end), and `f_bottom` (where the `Ds` end).

```python
d_bottom = np.add(c_bottom, Cs)
f_bottom = np.add(d_bottom, Ds)
```

2. Create a figure of width `10` and height `8`.

```python
%%capture
plt.figure(figsize=(10,8))
```

3. Plot the `As`, `Bs`, `Cs`, `Ds`, and `Fs`. Give each one the appropriate bottom list that will stack them on top of each other.

```python
plt.figure(figsize=(10,8))
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)
plt.show()
```

4. Create a set of axes and save them to `ax`.

```python
plt.figure(figsize=(10,8))
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)
ax = plt.subplot()
plt.show()
```

5. Set the x-ticks to be `range(len(unit_topics))`.

```python
plt.figure(figsize=(10,8))
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)
ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))
plt.show()
```

6. Set the x-tick labels to be the `unit_topics`.

```python
plt.figure(figsize=(10,8))
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)
ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)
plt.show()
```

7. Give the plot the title you see in the final graph, and the same x-axis label and y-axis label.

```python
plt.figure(figsize=(10,8))
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)
plt.title("Grade distribution", fontsize=25)
plt.xlabel("Unit", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)
plt.show()
```

8. Save your figure to a file called `my_stacked_bar.png`.

```python
plt.figure(figsize=(10,8))
plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)
plt.title("Grade distribution", fontsize=25)
plt.xlabel("Unit", fontsize=15)
plt.ylabel("Number of Students", fontsize=15)
ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)
plt.savefig("my_stacked_bar.png")
plt.show()
```

## Line with Shaded Error

Now, we are going to look at the chart called `line_graph.png`. This displays the relationship between the time students say they studied and the scores they received on their final exams.<br>
<img src="https://content.codecademy.com/courses/matplotlib/project_ii/fill_between.png" alt="fill between" style="background-color:white;"></img><br>
The data you will need to recreate this chart is in the lists `hours_reported` and `exam_scores`.<br>
<br>
Save your recreated chart to a file called `my_line_graph.png`.

```python
from matplotlib import pyplot as plt

hours_reported =[3, 2.5, 2.75, 2.5, 2.75, 3.0, 3.5, 3.25, 3.25,  3.5, 3.5, 3.75, 3.75,4, 4.0, 3.75,  4.0, 4.25, 4.25, 4.5, 4.5, 5.0, 5.25, 5, 5.25, 5.5, 5.5, 5.75, 5.25, 4.75]
exam_scores = [52.53, 59.05, 61.15, 61.72, 62.58, 62.98, 64.99, 67.63, 68.52, 70.29, 71.33, 72.15, 72.67, 73.85, 74.44, 75.62, 76.81, 77.82, 78.16, 78.94, 79.08, 80.31, 80.77, 81.37, 85.13, 85.38, 89.34, 90.75, 97.24, 98.31]
```

1. Create a figure of width `10` and height `8`.

```python
%%capture
plt.style.use('default')
plt.figure(figsize=(10,8))
```

2. Plot the `exam_scores` on the x-axis and the `hours_reported` on the y-axis, with a linewidth of `2`.

```python
plt.figure(figsize=(10,8))
plt.plot(exam_scores, hours_reported, linewidth=2)
plt.show()
```

3. Let us assume the error on students' reporting of their hours studying is 20%.<br>
<br>
Create a list `hours_lower_bound`, which has the elements of `hours_reported` minus 20%, and a list `hours_upper_bound`, which has the elements of `hours_reported` plus 20%.<br>
<br>
You can do this with a list comprehension that looks like:<br>
        
    y_lower_bound = [element - (element * error_in_decimal) for element in original_list_of_y_values]

```python
hours_lower_bound = [i - (i * 0.2) for i in hours_reported]
hours_upper_bound = [i + (i * 0.2) for i in hours_reported]

print(hours_lower_bound)
print(hours_upper_bound)
```

4. Shade the area between `hours_lower_bound` and `hours_upper_bound` with an alpha of `0.2`.

```python
plt.figure(figsize=(10,8))
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha=0.2)
plt.plot(exam_scores, hours_reported, linewidth=2)
plt.show()
```

5. Give the plot the title `'Time spent studying vs final exam scores'`, x-axis label `'Score'`, and y-axis label `'Hours studying (self-reported)'`.

```python
plt.figure(figsize=(10,8))
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha=0.2)
plt.plot(exam_scores, hours_reported, linewidth=2)
plt.title('Time spent studying vs final exam scores', fontsize=25)
plt.xlabel('Score', fontsize=15)
plt.ylabel('Hours studying (self-reported)', fontsize=15)
plt.show()
```

6. Save your figure to a file called `my_line_graph.png`.

```python
plt.figure(figsize=(10,8))
plt.fill_between(exam_scores, hours_lower_bound, hours_upper_bound, alpha=0.2)
plt.plot(exam_scores, hours_reported, linewidth=2)
plt.title('Time spent studying vs final exam scores', fontsize=25)
plt.xlabel('Score', fontsize=15)
plt.ylabel('Hours studying (self-reported)', fontsize=15)
plt.savefig('my_line_graph.png')
plt.show()
```
