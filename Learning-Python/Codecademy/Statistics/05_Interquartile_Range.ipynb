# Interquartile Range

## Range Review

One of the most common statistics to describe a dataset is the range. The range of a dataset is the difference between the maximum and minimum values. While this descriptive statistic is a good start, it is important to consider the impact outliers have on the results:

<img src="https://content.codecademy.com/courses/statistics/quantiles/outliers.svg" alt="A dataset with some outliers." style="background-color:white;"></img><br>

In this image, most of the data is between `0` and `15`. However, there is one large negative outlier (`-20`) and one large positive outlier (`40`). This makes the range of the dataset `60` (The difference between `40` and `-20`). Thatâ€™s not very representative of the spread of the majority of the data!

The _interquartile range_ (IQR) is a descriptive statistic that tries to solve this problem. The IQR ignores the tails of the dataset, so you know the range around-which your data is centered.

In this lesson, we will teach you how to calculate the interquartile range and interpret it.

### Exercise

```python
import json

with open('songs.json', 'r') as json_file:
  imported_data = json.load(json_file)

songs = imported_data['songs']
```

```python
import matplotlib.pyplot as plt

maximum = max(songs)
minimum = min(songs)

# Plot
plt.hist(songs, bins = 200)
plt.xlabel("Song Length (Seconds)")
plt.ylabel("Count")
plt.show()
```

1. We have imported a dataset of song lengths (measured in seconds) and plotted a histogram.

    It looks like there are some outliers â€” this might be a good opportunity to use the IQR.

    Before we do that, let us calculate the range. We have found the maximum and minimum values of the dataset and stored them in variables named `maximum` and `minimum`.

    Create a variable named `song_range` and set it equal to the difference between the maximum and the minimum.

```python
song_range = maximum - minimum
```

```python
# Run this cell once you are done
try:
  print("The range of the dataset is " + str(song_range) + " seconds")
except NameError:
  print("You haven't defined the variable song_range yet")
```

## Quartiles

The interquartile range is the difference between the third quartile (Q3) and the first quartile (Q1).

For now, all you need to know is that the first quartile is the value that separates the first 25% of the data from the remaining 75%.

The third quartile is the opposite â€” it separates the first 75% of the data from the remaining 25%.

<img src="https://content.codecademy.com/courses/statistics/quantiles/interquartile.svg" alt="The interquartile range of the dataset is shown to be between Q3 and Q1." style="background-color:white;"></img><br>

The interquartile range is the difference between these two values.

***
### Exercise

```python
import numpy as np

q1 = np.quantile(songs, 0.25)
```

1. We have calculated the first quartile of songs and stored it in the variable `q1`.

    Calculate the third quartile and store it in a variable named `q3`.

    To calculate the third quartile, call the same function, but change the second parameter to `0.75`.

```python
q3 = np.quantile(songs, 0.75)
```

2. Now that we have both the first quartile and the third quartile, let us calculate the IQR.

    Create a variable named `interquartile_range` and set it equal to the difference between `q3` and `q1`.

```python
interquartile_range = q3 - q1
```

```python
# Run this cell once you are done
try:
  print("The first quartile of the dataset is " + str(q1) + "\n")
except NameError:
  print("You haven't defined q1 yet\n")
  
try:
  print("The third quartile of the dataset is " + str(q3) + "\n")
except NameError:
  print("You haven't defined q3 yet\n")
  
try:
  print("The IQR of the dataset is " + str(interquartile_range) + "\n")
except NameError:
  print("You haven't defined interquartile_range yet\n")
```

## IQR in SciPy

In the last exercise, we calculated the IQR by finding the quartiles using NumPy and finding the difference ourselves. The SciPy library has a function that can calculate the IQR all in one step.

The `iqr()` function takes a dataset as a parameter and returns the Interquartile Range. Notice that when we imported `iqr()`, we imported it from the `stats` submodule:

    From scipy.stats import iqr
 
    dataset = [4, 10, 38, 85, 193]
    interquartile_range = iqr(dataset)

***

### Exercise

1. Let us calculate the IQR again, but this time, use SciPy's function.

    Create a variable named `interquartile_range` and set it equal to the result of calling `iqr()` using `songs` as a parameter.

```python
from scipy.stats import iqr

interquartile_range = iqr(songs)
```

```python
# Run this cell once you are done
try:
  print("The IQR of the dataset is " + str(interquartile_range) + "\n")
except NameError:
  print("You haven't defined interquartile_range yet\n")
```

## Review

Nice work! You can now calculate the Interquartile Range of a dataset by using the SciPy library. The main takeaway of the IQR is that it is a statistic, like the range, that helps describe the spread of the center of the data.

However, unlike the range, the IQR is robust. A statistic is robust when outliers have little impact on it. For example, the IQRs of the two datasets below are identical, even though one has a massive outlier.

    dataset_one = [6, 9, 10, 45, 190, 200] #IQR is 144.5
    dataset_two = [6, 9, 10, 45, 190, 20000000] #IQR is 144.5

By looking at the IQR instead of the range, you can get a better sense of the spread of the middle of the data.

The interquartile range is displayed in a commonly-used graph â€” the box plot.

<img src="https://content.codecademy.com/courses/statistics/quantiles/boxplot.png" alt="A box plot." style="background-color:white;"></img><br>

In a box plot, the ends of the box are Q1 and Q3. So the length of the box is the IQR. You can learn more about box plots in our upcoming course!
