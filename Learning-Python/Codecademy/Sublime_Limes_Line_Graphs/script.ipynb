## Understand the data and set up subplots

1. We have provided some data in different lists in the cell below. Look through these lists and try to understand what each one represents.

```python
from matplotlib import pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

2. Create a figure of width 12 and height 8.

```python
plt.figure(figsize=(12, 8))
```

Write the command to create the left subplot (the one that would correspond to the plot with a star in our example figure). Save this subplot in a variable called `ax1`.

4. Write the command to create the right subplot (the one that would correspond to the plot with a square in our example figure).<br>
<br>
Save this subplot in a variable called `ax2`.

```python
ax2 = plt.subplot(1, 2, 2)
```

## Page visits over time

5. In the left subplot, we are going to plot the total page visits over the past year as a line.<br>
<br>
First, let us create the list of x-values, which is `range(len(months))`. Store this in a variable called `x_values`.<br>
<br>
Make sure this happens after the line where you created `ax1`, but before the line where you have created `ax2`, so that the plot goes in the subplot on the left.

```python
x_values = range(len(months))
```

6. Plot the total page visits against these `x_values` as a line.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month)
ax2 = plt.subplot(1, 2, 2)
```

7. Give the line markers that will help show each month as a distinct value.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
ax2 = plt.subplot(1, 2, 2)
```

8. Label the x-axis and y-axis with descriptive titles of what they measure.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.ylabel('Visits')

ax2 = plt.subplot(1, 2, 2)
```

9. Set the x-axis ticks to be the `x_values`.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.ylabel('Visits')
plt.xticks(x_values)

ax2 = plt.subplot(1, 2, 2)
```

10. Label the x-axis tick labels to be the names stored in the `months` list.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

ax2 = plt.subplot(1, 2, 2)
```

## Plotting multiple lime species
11. In the subplot on the right, we are going to plot three lines on the same set of axes. The x-values for all three lines will correspond to the months, so we can use the list of `x_values` we used for the last plot.<br>
<br>
On one plot, create the three lines:<br>

* number of key limes sold vs `x_values`
* number of Persian limes sold vs `x_values`
* number of blood limes sold vs `x_values`
<br>
<br>
Make sure this happens after the line where you created ax2, so that it goes in the subplot on the right.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month)
plt.plot(x_values, persian_limes_per_month)
plt.plot(x_values, blood_limes_per_month)

plt.show()
```

12. Give each line a specific color of your choosing.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color='Lime')
plt.plot(x_values, persian_limes_per_month, color='Black')
plt.plot(x_values, blood_limes_per_month, color='Red')

plt.show()
```

13. Add a legend to differentiate the lines, labeling each lime species.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color='Lime')
plt.plot(x_values, persian_limes_per_month, color='Black')
plt.plot(x_values, blood_limes_per_month, color='Red')
plt.legend(['Key', 'Persian', 'Blood'])

plt.show()
```

14. Set the x-axis ticks to be the x_values, and the tick labels to be the `months` list.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color='Lime')
plt.plot(x_values, persian_limes_per_month, color='Black')
plt.plot(x_values, blood_limes_per_month, color='Red')
plt.xlabel('Months')
plt.xticks(rotation = 45)
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.legend(['Key', 'Persian', 'Blood'])

plt.show()
```

## Labeling and saving

15. Add a title to each of the two plots you have created, and adjust the margins to make the text you have added look better.

```python
ax1 = plt.subplot(1, 2, 1)
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.title('Visits per month')

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color='Lime')
plt.plot(x_values, persian_limes_per_month, color='Black')
plt.plot(x_values, blood_limes_per_month, color='Red')
plt.xlabel('Months')
plt.xticks(rotation = 45)
plt.ylabel('Number of limes')
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.legend(['Key', 'Persian', 'Blood'])
plt.title('Sales per month')


plt.subplots_adjust(wspace=0.35)

plt.show()
```
