# Traveling to Acadia

Imagine you work at a travel agency and want to inform your customers of the best time to visit one of your favorite vacation locations, Acadia, Maine. In this project, you will use flower bloom data, and flight information to recommend the best time of year for someone to make a trip to Maine.

In cell below, we have imported the following data:

* `in_bloom` is a NumPy array containing 900 numbers. Each number is a day of the year, which represents the average start date of a flower blooming.
* `flights` is a NumPy array with 11,000 numbers. Each number is a day of the year, which represents a flight from your hometown to airports near Acadia, Maine.

Your goal is to create two histograms, each displaying the frequency of an occurrence each day of the year (either flights or flower blooms).

You will use the `in_bloom` variable to find a count of the number of flowers that start blooming each day of the year.

You will use the `flights` variable to find a count of the number of flights that occur each day of the year.

```python
import numpy as np
from matplotlib import pyplot as plt

in_bloom = np.loadtxt(open("in-bloom.csv"), delimiter=",")
flights = np.loadtxt(open("flights.csv"), delimiter=",")
```

## Making a Histogram

1. Use Matplotlib to create a flights histogram.

```python
plt.hist(flights, bins=52, edgecolor='black')
plt.show()
```

2. If you have not done so already, set the range of your histogram to `(0, 365)`

```python
plt.hist(flights, bins=52, range=(0,365), edgecolor='black')
plt.show()
```

3. Set the number of bins in your plot to 365, so you have a separate bin for each day of the year.

```python
plt.hist(flights, bins=365, range=(0,365))
plt.show()
```

4. Add an x-label, y-label, and title to your figure.

```python
plt.hist(flights, bins=365, range=(0,365))
plt.title("Flight Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 
plt.show()
```

## Multiple Plots

5. Now, we are going to set up our figure so it displays two plots at once. Above your `plt.hist()` line, add the following:

        plt.figure(1)
        plt.subplot(211)

```python
plt.figure(1)

plt.subplot(211)
plt.hist(flights, bins=365, range=(0,365))
plt.title("Flight Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 

plt.show()
```

```python
plt.figure(1)

plt.subplot(211)
plt.hist(flights, bins=365, range=(0,365))
plt.title("Flight Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 

plt.subplot(212)

plt.show()
```

## Second Histogram

7. Under `plt.subplot(212)`, use `plt.hist()` to make a histogram that displays the number of flowers that begin to bloom each day of the year.

```python
plt.figure(1)

plt.subplot(211)
plt.hist(flights, bins=365, range=(0,365))
plt.title("Flight Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 

plt.subplot(212)
plt.hist(in_bloom, bins=365, range=(0,365))
plt.title("Flower Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 

plt.show()
```

8. Label the y-axis of the plot. In the hint, we have added our code.

    Right before calling `plt.show()` call `plt.tight_layout()`. This will prevent the labels from overlapping with the graphs.

```python
plt.figure(1)

plt.subplot(211)
plt.hist(flights, bins=365, range=(0,365))
plt.title("Flight Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 

plt.subplot(212)
plt.hist(in_bloom, bins=365, range=(0,365))
plt.title("Flower Frequency")
plt.xlabel("Day")
plt.ylabel("Count") 

plt.tight_layout()
plt.show()
```
