## Basic Line Plot

* `plt.show()` will actually display the graph

## Linestyles

We can specify a different color for a line by using the keyword color with either an <a href="https://www.w3schools.com/colors/colors_names.asp">HTML color</a> name or a <a href="https://www.w3schools.com/colors/colors_picker.asp">HEX code</a>:

```python
plt.plot(days, money_spent, color='green')
plt.plot(days, money_spent_2, color='#AAAAAA')
plt.show()
```

We can also make a line dotted or dashed using the keyword `linestyle`.

```python
# Dashed:
plt.plot(days, money_spent, linestyle='--')
plt.show()
```

```python
# Dotted:
plt.plot(days, money_spent, linestyle=':')
plt.show()
```

```python
# No line:
plt.plot(days, money_spent, linestyle='', marker='s')
plt.show()
```

We can also add a marker using the keyword `marker`:

```python
# A circle:
plt.plot(days, money_spent, marker='o')
plt.show()
```

```python
# A square:
plt.plot(days, money_spent, marker='s')
plt.show()
```

```python
# A star:
plt.plot(days, money_spent, marker='*')
plt.show()
```

To see all of the possible options, check out the <a href="https://matplotlib.org/api/lines_api.html">Matplotlib documentation</a>. Here are a couple of those values applied to our plots about lunch spending:

```python
plt.plot(days, money_spent, color='green', linestyle='--')
plt.plot(days, money_spent_2, color='#AAAAAA',  marker='o')
plt.show()
```

Sometimes, it can be helpful to zoom in or out of the plot, especially if there is some detail we want to address. To zoom, we can use `plt.axis()`. We use `plt.axis()` by feeding it a list as input. This list should contain:

* The minimum x-value displayed
* The maximum x-value displayed
* The minimum y-value displayed
* The maximum y-value displayed

For example, if we want to display a plot from `x=0` to `x=3` and from `y=2` to `y=5`, we would call `plt.axis([0, 3, 2, 5])`.

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.axis([0, 3, 2, 5])
plt.show()
```

## Subplots

Sometimes, we want to display two lines side-by-side, rather than in the same set of x- and y-axes. When we have multiple axes in the same picture, we call each set of axes a <i>subplot</i>. The picture or object that contains all of the subplots is called a <i>figure</i>.<br>
<br>
We can have many different subplots in the same figure, and we can lay them out in many different ways. We can think of our layouts as having rows and columns of subplots. For instance, the following figure has six subplots split into 2 rows and 3 columns:

```python
from IPython.display import Image
Image(url='https://content.codecademy.com/courses/matplotlib/six_subplots.svg', width=500)
```

We can create subplots using `.subplot()`.<br>
<br>
The command `plt.subplot()` needs three arguments to be passed into it:

* The number of rows of subplots
* The number of columns of subplots
* The index of the subplot we want to create

For instance, the command `plt.subplot(2, 3, 4)` would create "Subplot 4" from the figure above.<br>
<br>
Any `plt.plot()` that comes after `plt.subplot()` will create a line plot in the specified subplot. For instance:

```python
# Data sets
x = [1, 2, 3, 4]
y = [1, 2, 3, 4]
 
# First Subplot
plt.subplot(1, 2, 1)
plt.plot(x, y, color='green')
plt.title('First Subplot')
 
# Second Subplot
plt.subplot(1, 2, 2)
plt.plot(x, y, color='steelblue')
plt.title('Second Subplot')
 
# Display both subplots
plt.show()
```

## Subplots Part II

Sometimes, when we are putting multiple subplots together, some elements can overlap and make the figure unreadable:<br>
<br>
<img src="https://content.codecademy.com/courses/matplotlib/overlapping_subplots.png" alt="Plot with overlapping subplots" style="background-color:white;"></img><br>
<br>
We can customize the spacing between our subplots to make sure that the figure we create is visible and easy to understand. To do this, we use the `plt.subplots_adjust()` command. `.subplots_adjust()` has some keyword arguments that can move your plots within the figure:

* `left` â€” the left-side margin, with a default of 0.125. You can increase this number to make room for a y-axis label
* `right` â€” the right-side margin, with a default of 0.9. You can increase this to make more room for the figure, or decrease it to make room for a legend
* `bottom` â€” the bottom margin, with a default of 0.1. You can increase this to make room for tick mark labels or an x-axis label
* `top` â€” the top margin, with a default of 0.9
* `wspace` â€” the horizontal space between adjacent subplots, with a default of 0.2
* `hspace` â€” the vertical space between adjacent subplots, with a default of 0.2

For example, if we were adding space to the bottom of a graph by changing the bottom margin to 0.2 (instead of the default of 0.1), we would use the command:<br>
<br>
`plt.subplots_adjust(bottom=0.2)`<br>
<br>
We can also use multiple keyword arguments, if we need to adjust multiple margins. For instance, we could adjust both the top and the hspace:<br>
<br>
`plt.subplots_adjust(top=0.95, hspace=0.25)`<br>
<br>
Let us use `wspace` to fix the figure above:

```python
# Left Plot
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])
plt.xlabel('x-label for Plot 1')
plt.ylabel('y-label for Plot 1')

# Right Plot
plt.subplot(1, 2, 2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])
plt.xlabel('x-label for Plot 2')
plt.ylabel('y-label for Plot 2')

# Subplot Adjust
plt.subplots_adjust(wspace=0.35)
 
plt.show()
```

## Legends

When we have multiple lines on a single graph we can label them by using the command `plt.legend()`.<br>
<br>
The `legend` method takes a list with the labels to display. So, for example, we can call:

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
plt.legend(['parabola', 'cubic'])
plt.show()
```

This displays a legend on our graph, labeling each line.<br>
<br>
`plt.legend()` can also take a keyword argument `loc`, which will position the legend on the figure.<br>
<br>
These are the position values `loc` accepts:

|Number Code|String|
|:----------|:-----|
|0|best|
|1|upper right|
|2|upper left|
|3|lower left|
|4|lower right|
|5|right|
|6|center left|
|7|center right|
|8|lower center|
|9|upper center|
|10|center|

<b>Note:</b> If you decide not to set a value for loc, it will default to choosing the "best" location.<br>
<br>
For, example, we can call `plt.legend()` and set `loc` to `6`:

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
plt.legend(['parabola', 'cubic'], loc=6)
plt.show()
```

This moves the legend to the left side of the graph.<br>
<br>
Sometimes, it is easier to label each line as we create it. If we want, we can use the keyword `label` inside of `plt.plot()`. If we choose to do this, we do not pass any labels into `plt.legend()`. For example:

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16],
         label="parabola")
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64],
         label="cubic")
plt.legend() # Still need this command!
plt.show()
```

This displays a legend that looks just like what we had before.

## Modify Ticks

In all of our previous exercises, our commands have started with `plt.`. In order to modify tick marks, we will have to try something a little bit different.<br>
<br>
Because our plots can have multiple subplots, we have to specify which one we want to modify. In order to do that, we call `plt.subplot()` in a different way.<br>
<br>
`ax = plt.subplot(1, 1, 1)`<br>
<br>
`ax` is an axes object, and it lets us modify the axes belonging to a specific subplot. Even if we only have one subplot, when we want to modify the ticks, we will need to start by calling either `ax = plt.subplot(1, 1, 1)` or `ax = plt.subplot`() in order to get our axes object.<br>
<br>
Suppose we wanted to set our x-ticks to be at 1, 2, and 4. We would use the following code:

```python
ax = plt.subplot()
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])
ax.set_xticks([1, 2, 4])
plt.legend(['parabola', 'cubic'])
plt.show()
```

We can also modify the y-ticks by using `ax.set_yticks()`.<br>
<br>
When we change the x-ticks, their labels automatically change to match. But, if we want special labels (such as strings), we can use the command `ax.set_xticklabels()` or `ax.set_yticklabels()`. For example, we might want to have a y-axis with ticks at 0.1, 0.6, and 0.8, but label them 10%, 60%, and 80%, respectively. To do this, we use the following commands:

```python
ax = plt.subplot()
plt.plot([1, 3, 3.5], [0.1, 0.6, 0.8], 'o')
ax.set_yticks([0.1, 0.6, 0.8])
ax.set_yticklabels(['10%', '60%', '80%'])
plt.show()
```

## Figures

When we are making lots of plots, it is easy to end up with lines that have been plotted and not displayed. If we are not careful, these "forgotten" lines will show up in your new plots. In order to be sure that you do not have any stray lines, you can use the command `plt.close('all')` to clear all existing plots before you plot a new one.<br>
<br>
Previously, we learned how to put two sets of axes into the same figure. Sometimes, we would rather have two separate figures. We can use the command `plt.figure()` to create new figures and size them how we want. We can add the keyword `figsize=(width, height)` to set the size of the figure, in inches. We use parentheses (`(` and `)`) to pass in the width and height, which are separated by a comma (`,`).<br>
<br>
To create a figure with a width of 4 inches, and height of 10 inches, we would use:

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.figure(figsize=(4, 10))
plt.plot(x, y)
plt.show()
```

Once we have created a figure, we might want to save it so that we can use it in a presentation or a website. We can use the command `plt.savefig()` to save out to many different file formats, such as `png`, `svg`, or `pdf`. After plotting, we can call `plt.savefig('name_of_graph.png')`:

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.figure(figsize=(4, 10)) 
plt.plot(x, y)
plt.savefig('tall_and_narrow.png')
```
